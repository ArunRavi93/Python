### Cutomizing pie plot in matplot lib

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 
data = [23, 17, 35, 29, 12, 41]
 
 
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```

## Seaborn Categorical Plots

```python
# barplot: It is a general plot that helps us to aggregate the data
# bar plot is a general plot that allos you to aggregate the categorical data based of some
# function, by default it is the mean function
# Categorical vs Continuous

sns.barplot(x='sex',y='total_bill',data=tips); # xaxis categorical colum # y axis = numerical
```

```python
import numpy as np
# estimator is basically the function to be applied
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std); # here the estimator is std Dev
```

```python
# countplot: It is same as barplot except the estimator here is explicitly counting
# the no of occurences. by default the estimator here is count
# Therefore we only pass x value

sns.countplot(x='sex',data=tips); # here we need not input y axis because it is always count
```

# Scale and Context

> set_context : seaborn.set_context(context=None, font_scale=1, rc=None)
> Set the parameters that control the scaling of plot elements.

> This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. This is accomplished using the matplotlib rcParams system.

> Ex : poster, notebook, paper, talk
